```python
import pandas as pd
```

# Read csv file as dataframe

```python
df=pd.read_csv("/content/takehome_users.csv",encoding='utf-8', encoding_errors='ignore')
df1=pd.read_csv("/content/takehome_user_engagement.csv")
```

```python
df1
```

# seperate date from datafreame and insert into new column

```python
df1["time"] = pd.to_datetime(df1['time_stamp']).dt.date
df1
```

# Find unique date and remove same date from dataframe

```python
q=df1.time.unique()
```

```python
q
```

# Get pre-defined unique date value of data's from dataframe

```python
df2=(df1.loc[df1['time'].isin(q)])
df2
```

# Convert date to weeks and insert into new column

```python
week_val= pd.to_datetime(df2["time"])
result = week_val.dt.week
result
```

```python
df2.insert(1, "week_no", result)
df2
```

```python
df2['year'] = pd.DatetimeIndex(df2['time']).year
df2
```

# **Get year and month in seperate column**

```python
import datetime
```

```python
df2['y_m']= list(zip(df2['week_no'], df2['year']))
```

```python
df2.head(20)
```

```python
 data=df2.groupby(["user_id","y_m"])["visited"].count().reset_index()
 data.head(20)
```

# Find adopted user who visites greater than or equal to the times

```python
adopted_user=data.loc[data["visited"]>=3]
adopted_user.head(20)
```

```python
data['adopted_user'] = data['visited'].apply(lambda x:"adopted_user" if x>=3 else "non adopted_user")
```

```python
data
```

```python
df.rename(columns={"object_id":"user_id"},inplace =True)
df.head()
```

```python
df
```

# Merge two data frame for more details about adopted users 

```python
merged_data=pd.merge(df,data,how="right",on="user_id")
```

```python
merged_data
```

# Visualize adopted user 

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
x = ['Adopted_users','Non_adopted_users']
y = [(data.adopted_user.value_counts()[1]),(data.adopted_user.value_counts()[0])]
plt.bar(x,y,color='g',edgecolor='r',width=0.5)

plt.xlabel("Type of users",color='blue')
plt.ylabel("Number of users",color='blue')
plt.title("Visualisation on type of users",color='m',fontsize=14)
plt.show()
```
